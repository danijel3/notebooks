```python
import cv2
import urllib
import numpy as np

import matplotlib.pyplot as P
%matplotlib inline

#req = urllib.urlopen('http://i.imgur.com/QSMu4W3.jpg')
#req = urllib.urlopen('http://i.imgur.com/v4NIfCo.jpg')
#req = urllib.urlopen('http://i.imgur.com/O1Tme5a.jpg')
req = urllib.urlopen('http://i.imgur.com/0qnICGP.jpg')
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
img = cv2.imdecode(arr,-1) # 'load it as it is'

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

P.imshow(img)

print img.shape
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

center_x=hsv.shape[0]/2
center_y=hsv.shape[1]/2

sq=hsv[center_x-10:center_x+10,center_y-10:center_y+10]

h,s,v = cv2.split(sq)

h = np.median(h)
s = np.median(s)
v = np.median(v)

HRANGE=10
SRANGE=100
VRANGE=100

lower = np.array([h-HRANGE,s-SRANGE,v-VRANGE])
upper = np.array([h+HRANGE,s+SRANGE,v+VRANGE])

mask = cv2.inRange(hsv, lower, upper)

kernel = np.ones((10,10),np.uint8)
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

res = cv2.bitwise_and(img,img, mask=mask)

P.imshow(res,cmap=P.cm.gray)
```
